```python
import matplotlib.pyplot as plt
from numpy import meshgrid, arange, shape, sin, cos
```

```python
%matplotlib inline
```

```python
x, y = meshgrid(arange(-100,100),arange(-100,100))
```

```python
data = sin((x/5.0)**2+(y/5.0)**2)
```

```python
plt.figure(figsize = (20,20))
plt.imshow(data,cmap='gray')
```
