```python
from sympy import *
```

```python
init_printing()
```

```python
g = symbols('gamma')
w = symbols('omega')
w0 = symbols('omega_0')
dn = symbols('delta_n')
```

```python
subs = {dn:1, w0:115, g:2}
```

```python
n_real = 1 + dn*(2*(w0 - w)*g)/((w0 - w)**2 + g**2)
```

```python
n_imag = dn*(g**2)/((w0-w)**2 + g**2)
```

```python
ng = n_real + w*diff(n_real,w)
```

```python
plot(n_real.subs(subs),(w,100,125))
plot(n_imag.subs(subs),(w,100,125))
plot(ng.subs(subs),(w,100,125))
```
