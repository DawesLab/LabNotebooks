```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
theta_max = 90
degrees = np.linspace(0,theta_max)
radians = np.radians(degrees)
```

```python
def N(theta):
    N = mg - F*np.sin(theta)
    if N.any() < 0:
    return N
```

```python
F = 5
mg = 1
def work(mu, theta):
    W = mu*N(theta) + F*np.cos(theta)
    return W
```

```python
mulist = np.linspace(0.1,0.8,10)
for mu in mulist:
    plt.plot(degrees,work(mu,radians))
```
