```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as sp
import scipy.signal as ss
from numpy.random import default_rng

rng = default_rng()
N = 500
x = 0.1*rng.standard_normal(N)
y = rng.standard_normal(N)
```

```python
t = np.linspace(0,200,N)
```

```python
sigA = np.sin(0.34*t) + x
sigB = np.sin(0.34*t + 1.2) + x
```

```python
sigA = y + x
sigB = np.roll(y,15) + x
```

```python
plt.subplot(121)
plt.plot(sigA)
plt.plot(sigB)
plt.subplot(122)
plt.plot(sigA,sigB)
```

```python
plt.plot(ss.correlate(sigA,sigB)[450:550])
```

```python
phase = np.linspace(0,2*np.pi,40)
sigA = np.sin(0.34*t) + x
r_values = []
for p in phase: 
    sigB = np.sin(0.34*t + p) + x
    Pr, Pp = sp.pearsonr(sigA,sigB)
    r_values.append(Pr)
    
```

```python
plt.plot(phase,r_values)
```

```python
ss.correlate(sigA,sigB).max()
```
